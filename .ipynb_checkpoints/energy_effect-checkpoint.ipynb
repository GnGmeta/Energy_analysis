```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
%matplotlib inline
sns.set_style('dark')
```

```python
df = pd.read_csv('energy.csv')
```

```python
df = df.iloc[:,0:4]
```

```python
df_time = df.unixtimestamp.apply(datetime.fromtimestamp)
```

```python
year = df_time.astype('str').str.slice(0,4)
month = df_time.astype('str').str.slice(5,7)
day = df_time.astype('str').str.slice(8,10)
hour = df_time.astype('str').str.slice(11,13)
minute = df_time.astype('str').str.slice(14,16)
weekday = df_time.apply(datetime.weekday)
df['timestamp'] = df_time
df['year'] = year
df['month'] = month
df['day'] = day
df['hour'] = hour
df['minute'] = minute
df['weekday'] = weekday
```

```python
df_clean = df.loc[df.max_demand != 0, :]
df_clean = df_clean.loc[df_clean.max_demand <= 75]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.distplot(df_clean['max_demand'].dropna(), ax=ax, kde=False).set_title('Max Demand', fontsize=16)
plt.xlabel('KW Power', fontsize=15)
plt.legend(['max demand'])
plt.title('Distribution of energy demanded ')
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(15,15))
for i in range(2):
    axs[i].plot(df_clean.groupby(['year','hour']).mean().max_demand.values)
    axs[i].set_xlabel('Hour of Year')
    axs[i].set_ylabel('Max Demand')
    axs[i].set_title('Year of ' +str(2018+i))
    plt.subplots_adjust(hspace=0.5)
    
```

```python
fig, axs = plt.subplots(12,1,figsize=(20,30))
for i in range(12):
    if i < 10:
        month = '0'+str(i+1)
    else:
        month = str(i+1)
    if i % 3 == 0:
        color = 'r'
    elif i % 3 == 1:
        color = 'g'
    else:
        color = 'b'
    axs[i].set_title('Month: '+month)
    dt = df_clean.loc[df_clean.month==month,'max_demand']
    axs[i].plot(dt,color=color)
    fig.show()
```

```python
def energy_save(effect):
    x = df_clean.groupby('month').mean().max_demand
    y = df_clean.groupby('month').mean().max_demand * (1-effect)
    plt.figure(figsize=(15,8))
    plt.plot(x,linewidth='4')
    plt.plot(y,linewidth='8')
    plt.xlabel('Month',fontsize=14)
    plt.ylabel('KWh',fontsize=15)
    plt.text('04',10.3,'Saved Power',fontsize=20,color='g')
    plt.annotate('Before',xy=('07',13.5),xytext=('06',14),
                arrowprops=dict(facecolor='black',shrink=0.06,),fontsize=14)
    plt.annotate('After',xy=('07',12),xytext=('06',10.5),
                arrowprops=dict(facecolor='black',shrink=0.06,),fontsize=14)
    plt.show() 
    
```

```python
energy_save(.12)
```
